# Data Preprocessing

```python
data.dtypes
```

### Features 'groupId', 'matchId', 'matchtype' are of categorical type 

```python
data.isna().sum()
```

### Output feature 'winPlacePerc' has a null value, so have to remove the corresponding row

```python
data[data.winPlacePerc.isna()]
```

```python
# removing above row 
data.drop(data[data["winPlacePerc"].isna()].index.values, inplace=True)
```

```python
anomaly = data[(data.kills == 0) & (data.winPlacePerc == 1) & (data.walkDistance == 0)]
anomaly
```

### We are looking for the players who have won the match, with zero walk distance but zero kills. This is the anomaly and we are removing it.

```python
clean_data = data.drop(index=anomaly.index)
```

# Basic Data Analysis

```python
print(f"Total number of players: {clean_data.shape[0]}")
print(f"Total number of players with winPlacePerc < 0.5: {clean_data[clean_data.winPlacePerc < 0.5].shape[0]}")
print(f"Percent of players with winPlacePerc < 0.5: {clean_data[clean_data.winPlacePerc < 0.5].shape[0]/clean_data.shape[0]}")
```

### We can see that out of 31 lakh players 16.5 lakh players have winPlacePerc < 0.5 and 15 lakh players have winPlacePerc > 0.5 therefore we can say that we have a almost balanced dataset

```python
def snsSize(x, y):
    sns.set(rc = {'figure.figsize':(x,y)})
def plotSize(x, y):
    plt.figure(figsize=(x,y))
```

```python
sns.scatterplot(x=clean_data.matchType, y=clean_data.kills)
snsSize(10, 20)
```

From the above scatter plot we can say that, the games of type **(normal-duo-fpp, normal-squad-fpp,normal-solo-fpp,normal-duo, normal-squad,normal-solo)** have a high kill range than the games of type **(solo-fpp, duo, squad, squad-fpp, duo-fpp, solo)**, which is much lower in games of type **(crashfpp, flarefpp,flaretpp,crashtpp)**

```python
sns.scatterplot(x=clean_data.matchType, y=clean_data.matchDuration)
snsSize(20, 10)
```

From the above scatter plot we can say that, the games of type **(solo-fpp, duo, squad, squad-fpp, duo-fpp, solo)** have a minimum threshold for matchduration but there is no such indication for the other types of games.

```python
sns.scatterplot(x=clean_data.matchType, y=clean_data.revives)
sns.set(rc = {'figure.figsize':(20,10)})
```

From the above scatter plot we can say that, the games of type **(duo, squad, squad-fpp, duo-fpp)** 
, have a higher number of revives made than games of type **(normal-duo-fpp, normal-squad-fpp,fpp,normal-duo, normal-squad)** which is much lower in games of type **(crashfpp, flarefpp,flaretpp,crashtpp)**



```python
sns.scatterplot(x=clean_data.matchType, y=clean_data.rideDistance)
sns.set(rc = {'figure.figsize':(20,10)})
```
